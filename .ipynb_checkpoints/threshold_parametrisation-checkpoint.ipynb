# Define the MBRL Agent 

```python
def DKL(d1,d2):
    if type(d1) != np.array and type(d2) != np.array:
        d1_array = np.array([d1[-1], d1[0], d1[1]])
        d2_array = np.array([d2[-1], d2[0], d2[1]])
        return np.sum(d1_array*np.log(d1_array/d2_array))
    
    elif type(d1) == np.array and type(d2) != np.array:
        d2_array = np.array([d2[-1], d2[0], d2[1]])
        return np.sum(d1*np.log(d1/d2_array))
    
    elif type(d2) == np.array and type(d1) != np.array:
        d1_array = np.array([d1[-1], d1[0], d1[1]])
        return np.sum(d1_array*np.log(d1_array/d2)) 
    
    else : 
        return np.sum(d1*np.log(d1/d2))

def swJS(d1,d2, N1, N2):
    d1_array = np.array([d1[-1], d1[0], d1[1]])
    d2_array = np.array([d2[-1], d2[0], d2[1]])
    return .5*DKL(d1_array,(N1*d1_array+N2*d2_array)/(N1+N2))+.5*DKL(d2_array,(N1*d1_array+N2*d2_array)/(N1+N2))

def input_dilemma(dilemma, Contexts, list_of_actions, threshold):
    reward_distribution = distribution(dilemma['Reward'])      #calculate the reward distribution 
    
    #Is the action performed in this dilemma already in the list of action ? 
    #If yes : 
    if dilemma['Action'] in list_of_actions :
        #We will iterate over the contexts already in memory
        context_list = list(Contexts[dilemma['Action']].keys())  
        DKL_context = [] 
        for i in range(len(context_list)):
            DKL_context.append(DKL(reward_distribution, Contexts[dilemma['Action']]['C'+str(i+1)]['Distribution'] )) #Caclulate the DKL between the dilemma distribution and the contexts' ones
        min_dkl, context_min_dkl = np.min(DKL_context), np.argmin(DKL_context) #Get the context for which the DKL is minimum
    
        if min_dkl > threshold : 
            Contexts[dilemma['Action']]['C'+str(1+len(Contexts[dilemma['Action']]))]={
                'Distribution' : reward_distribution,
                'Outcomes' : dilemma['Reward'],
                'States' : [[dilemma['State'], dilemma['Reward']]],
            }   #If the min DKL > threshold, we create a new context

    
        else : #Else, the current observation is added to an existing context
            Contexts[dilemma['Action']]['C'+str(context_min_dkl+1)]['Outcomes'] = np.concatenate((Contexts[dilemma['Action']]['C'+str(context_min_dkl+1)]['Outcomes'],dilemma['Reward']))    
            Contexts[dilemma['Action']]['C'+str(context_min_dkl+1)]['States'].append([dilemma['State'], dilemma['Reward']])
            Contexts[dilemma['Action']]['C'+str(context_min_dkl+1)]['Distribution'] = distribution(Contexts[dilemma['Action']]['C'+str(context_min_dkl+1)]['Outcomes'])
        
    #If the action performed is not in the list of action, we add it to the list and create a new context from it. 
    else :
        list_of_actions.append(dilemma['Action'])
        Contexts[dilemma['Action']] = {
            'C1':{'Distribution' : reward_distribution,
                'Outcomes' : dilemma['Reward'],
                'States' : [[dilemma['State'], dilemma['Reward']]],
            }
        }
        
def merge(Contexts, threshold):
    action_list = list(Contexts.keys())
    
    #Iterate over actions
    for action in action_list : 
        #We will calculate the swJS divergence for each pair of context for each action
        swJS_list = [[],[]]
        if len(Contexts[action]) > 1 : 
            for i in range(len(Contexts[action])):
                for j in range(i+1, len(Contexts[action])):
                    swJS_list[0].append([i,j])
                    swJS_list[1].append(swJS(Contexts[action]['C'+str(i+1)]['Distribution'], Contexts[action]['C'+str(j+1)]['Distribution'], len(Contexts[action]['C'+str(i+1)]['Outcomes']), len(Contexts[action]['C'+str(j+1)]['Outcomes']) ))
            #Take the min value of the swJS divergence for one action 
            min_swJS, context_min_swJS = np.min(swJS_list[1]), np.argmin(swJS_list[0])
            #Compare it to the threshold 
            if min_swJS < threshold : 
                i = swJS_list[0][context_min_swJS][0]
                j = swJS_list[0][context_min_swJS][1]
                n = len(Contexts[action])
                #Merge the two contexts
                Contexts[action]['C'+str(i+1)]['Outcomes'] = np.concatenate((Contexts[action]['C'+str(i+1)]['Outcomes'],Contexts[action]['C'+str(j+1)]['Outcomes']))
                Contexts[action]['C'+str(i+1)]['Outcomes'] = Contexts[action]['C'+str(i+1)]['Outcomes'].tolist()
                Contexts[action]['C'+str(i+1)]['States'] = np.concatenate((Contexts[action]['C'+str(i+1)]['States'],Contexts[action]['C'+str(j+1)]['States']))
                Contexts[action]['C'+str(i+1)]['States'] = Contexts[action]['C'+str(i+1)]['States'].tolist()
                Contexts[action]['C'+str(i+1)]['Distribution'] = distribution(Contexts[action]['C'+str(i+1)]['Outcomes'])
                #Replace the second one by the n^th context of the MBRL agent 
                Contexts[action]['C'+str(j+1)]['Outcomes'] = Contexts[action]['C'+str(n)]['Outcomes']
                Contexts[action]['C'+str(j+1)]['States'] = Contexts[action]['C'+str(n)]['States']
                Contexts[action]['C'+str(j+1)]['Distribution'] = Contexts[action]['C'+str(n)]['Distribution']
                #Delete the last context. It ensures that the contexts go from 1 to N 
                del Contexts[action]['C'+str(n)]
    
    return Contexts

def MBRL_agent(dilemma_list, Contexts, list_of_actions, adding_threshold, merging_threshold):
    for dilemma in dilemma_list : 
        input_dilemma(dilemma, Contexts, list_of_actions, adding_threshold)
        merge(Contexts, merging_threshold)          
    return Contexts 
    #return print_summary(Contexts)
    
```

# Plotting Functions 

```python
def normalize_distribution(distribution):
    """
    Normalize a ternary distribution so that the probabilities sum to 1.

    Args:
        distribution (dict): Dictionary with keys -1, 0, and 1 representing judgments.

    Returns:
        dict: Normalized distribution with values rounded to 4 decimals.
    """
    total = sum(distribution.values()) if distribution else 1
    return {k: round(v / total, 4) for k, v in distribution.items()}

def plot_distribution(distribution, title):
    """
    Display a simple bar chart for a ternary distribution (-1, 0, 1).

    Args:
        distribution (dict): Normalized distribution to be plotted.
        title (str): Title of the chart (e.g., Action - Context).
    """
    labels = ['Blame', 'Neutral', 'Support']
    values = [distribution.get(-1, 0), distribution.get(0, 0), distribution.get(1, 0)]

    plt.figure(figsize=(4, 3))
    plt.bar(labels, values, color=['red', 'gray', 'green'])
    plt.xlabel("Outcome")
    plt.ylabel("Probability")
    plt.title(title)
    plt.ylim(0, 1)
    plt.grid(axis="y", linestyle="--", alpha=0.6)
    # plt.show() is commented out to allow grouped display later

def print_summary(contexts):
    """
    Display a visual summary of the ternary distributions for each (action, context) pair.

    Args:
        contexts (dict): Nested dictionary of the form {action: {context: details}},
                         where details contains 'Distribution' (and optionally 'States').
    """
    table_data = []
    
    for action, context_dict in contexts.items():
        for context, details in context_dict.items():
            # Normalize the distribution, or use a zero-default distribution if missing
            normalized_dist = normalize_distribution(details['Distribution']) if details['Distribution'] else { -1: 0, 0: 0, 1: 0 }

            # Optional: text-based summary (currently commented)
            # dist_str = f"(-1: {normalized_dist[-1]:.2f}, 0: {normalized_dist[0]:.2f}, 1: {normalized_dist[1]:.2f})"
            # table_data.append([action, context, ', '.join(details['States'][0]) if details['States'][0] else "Ã˜", dist_str])
            
            # Plot one distribution per (action, context)
            plot_distribution(normalized_dist, title=f"{action} - {context}")

    # Save all figures as a single PDF (optional)
    plt.savefig('contexts_created_optimal_conditions.pdf')
    plt.show()

def distribution(samples):
    """
    Compute a ternary probability distribution from a list of sample values.

    Args:
        samples (list or array): List of sample values (e.g., [1, -1, 0, 1, 1, ...]).

    Returns:
        dict: Normalized probability distribution over -1, 0, and 1.
    """
    # Count occurrences of each unique value
    unique, counts = np.unique(samples, return_counts=True)

    # Compute relative frequencies
    total = len(samples)
    distribution = {val: count / total for val, count in zip(unique, counts)}

    # Ensure all three values (-1, 0, 1) are present with a minimum probability
    for val in [-1, 0, 1]:
        if val not in distribution:
            distribution[val] = 1e-5  # Small pseudo-count to avoid missing keys

    # Normalize to ensure total probability sums to 1
    total_prob = sum(distribution.values())
    distribution = {key: value / total_prob for key, value in distribution.items()}

    return distribution
```

# Definition of the loss function 

```python
# ==============================
#  Reference distributions 
# ==============================

ref_distrib = []
for name in base_distributions:
    ref_distrib.append(base_distributions[name])

# ==============================
# ðŸ“ Divergence & Distance Tools
# ==============================

def jsD(d1, d2):
    """
    Compute the Jensen-Shannon Divergence between two ternary distributions.

    Parameters:
        d1, d2: Lists or arrays representing distributions over [-1, 0, 1]

    Returns:
        Float: Jensen-Shannon Divergence
    """
    d1_array = np.array([d1[-1], d1[0], d1[1]])  # Reorder to match [-1, 0, 1]
    d2_array = np.array([d2[-1], d2[0], d2[1]])
    m = (d1_array + d2_array) / 2
    return 0.5 * jensenshannon(d1_array, m, base=2) ** 2 + 0.5 * jensenshannon(d2_array, m, base=2) ** 2


def emd(d1, d2):
    """
    Compute a simple Earth Moverâ€™s Distance (EMD) between two ternary distributions.

    Parameters:
        d1, d2: Lists or arrays representing distributions over [-1, 0, 1]

    Returns:
        Float: Earth Mover's Distance
    """
    d1_array = np.array([d1[-1], d1[0], d1[1]])
    d2_array = np.array([d2[-1], d2[0], d2[1]])
    
    # Cumulative distribution function method
    cdf_d1 = np.cumsum(d1_array)
    cdf_d2 = np.cumsum(d2_array)
    
    # Ground distance vector (|i - j| summed along rows)
    D = np.array([0, 1, 2])
    
    return np.sum(np.abs(cdf_d1 - cdf_d2) * D) / np.sum(D)

# ===========================================
# ðŸ§© Hungarian Matching for Distribution Sets
# ===========================================

def hungarian(predicted_distributions, reference_distributions, metric='emd'):
    """
    Match predicted distributions to reference ones using the Hungarian algorithm
    based on a chosen divergence metric.

    Parameters:
        predicted_distributions: List of predicted distributions
        reference_distributions: List of reference (canonical) distributions
        metric: Either 'emd' or 'jsD'

    Returns:
        Float: Total matching cost
    """
    num_pred = len(predicted_distributions)
    num_ref = len(reference_distributions)
    cost_matrix = np.zeros((num_pred, num_ref))

    for i in range(num_pred):
        for j in range(num_ref):
            if metric == 'emd':
                cost_matrix[i, j] = emd(predicted_distributions[i], reference_distributions[j])
            elif metric == 'jsD':
                cost_matrix[i, j] = jsD(predicted_distributions[i], reference_distributions[j])
    
    row_ind, col_ind = linear_sum_assignment(cost_matrix)
    return cost_matrix[row_ind, col_ind].sum()

# ==================================
# ðŸš¨ Penalized EMD with Duplicates
# ==================================

def penalized_emd(set_A, set_B, penalty_weight):
    """
    Compute EMD between two sets of distributions with a penalty for reusing references.

    Parameters:
        set_A: Predicted clusters (list of distributions)
        set_B: Reference canonical distributions
        penalty_weight: Weight of the penalty term for multiple assignments

    Returns:
        Float: Penalized EMD
    """
    support = np.array([-1, 0, 1])
    
    # Compute pairwise EMD matrix
    emd_matrix = np.array([
        [wasserstein_distance(support, support, a, b) for b in set_B]
        for a in set_A
    ])
    
    # Greedy matching: each predicted distribution chooses the closest reference
    pairing = [np.argmin(row) for row in emd_matrix]
    
    # Base cost
    base_cost = np.sum([emd_matrix[i][pairing[i]] for i in range(len(pairing))])
    
    # Penalize if the same reference is used multiple times
    _, counts = np.unique(pairing, return_counts=True)
    penalty = penalty_weight * np.sum(counts[counts > 1] - 1)
    
    return base_cost + penalty

# =====================================
# ðŸ§® Homogeneity of Cluster Assignments
# =====================================

def homogeneity(contexts):
    """
    Compute average homogeneity across clusters, based on dominant state frequencies.

    Parameters:
        contexts: Dictionary containing clusters with associated states

    Returns:
        Float: Average homogeneity score (1.0 = perfect homogeneity)
    """
    homog = 0
    for c in contexts['test']:
        L = [state[0] for state in contexts['test'][c]['States']]
        counter = Counter(L)
        dominant_state, count = counter.most_common(1)[0]
        homog += count / len(L) / len(contexts['test'])
    return homog

# ============================
# ðŸ’¡ Clustering Loss Functions
# ============================

def clustering_loss(params):
    """
    Clustering loss using Hungarian matching and cardinality penalty.

    Parameters:
        params: Tuple of (add_threshold, merge_threshold)

    Returns:
        Float: Total loss
    """
    add_thresh, merge_thresh = params
    Contexts = {}
    list_of_actions = []
    Contexts_opt = MBRL_agent(filtered_data, Contexts, list_of_actions, add_thresh, merge_thresh)
    
    clusters = [Contexts_opt['test'][c]['Distribution'] for c in Contexts_opt['test']]
    
    divergence = hungarian(ref_distrib, clusters, metric='emd')
    lambda_penalty = np.std(divergence)  # Optional: control cardinality deviation
    
    loss = divergence + lambda_penalty * abs(len(clusters) - len(ref_distrib))
    loss += np.random.normal(0, 0.01)  # Optional noise
    return loss

def clustering_loss_2(params):
    """
    Loss based on penalized EMD between predicted and reference clusters.

    Parameters:
        params: Tuple of (add_threshold, merge_threshold)

    Returns:
        Float: Penalized loss
    """
    add_thresh, merge_thresh = params
    Contexts = {}
    list_of_actions = []
    Contexts_opt = MBRL_agent(filtered_data, Contexts, list_of_actions, add_thresh, merge_thresh)
    
    clusters = [Contexts_opt['test'][c]['Distribution'] for c in Contexts_opt['test']]
    formatted = [[d[-1], d[0], d[1]] for d in clusters]
    
    return penalized_emd(formatted, ref_distrib, penalty_weight=0.6)

def clustering_loss_3(params):
    """
    Loss combining penalized EMD, context count, and homogeneity.

    Parameters:
        params: Tuple of (add_threshold, merge_threshold)

    Returns:
        Tuple: (loss, number of contexts, average homogeneity)
    """
    add_thresh, merge_thresh = params
    Contexts = {}
    list_of_actions = []
    Contexts_opt, nb_contexts = MBRL_agent_2(filtered_data, Contexts, list_of_actions, add_thresh, merge_thresh)
    
    clusters = [Contexts_opt['test'][c]['Distribution'] for c in Contexts_opt['test']]
    formatted = [[d[-1], d[0], d[1]] for d in clusters]
    
    loss = penalized_emd(formatted, ref_distrib, penalty_weight=0.6)
    homog = homogeneity(Contexts)
    
    return loss, nb_contexts, homog

def create_dist_list(N, S_list):
    # List of dictionaries
    distribution_dicts = []

    # Sample size
    num_samples = [11,30,50,100]  
    values = np.array([-1,0,1])


    for dist in base_distributions : 
        for i in range(len(S_list)):
            for j in range(N):
                samples = np.random.choice(values, size=S_list[i], p= base_distributions[dist])
            
                distribution_dicts.append({
                'Action': 'test',
                'State': dist+' '+str(S_list[i]),  
                'Reward': samples.tolist()  
                })


    random.shuffle(distribution_dicts)
    return(distribution_dicts)

def MBRL_agent_2(dilemma_list, Contexts, list_of_actions, adding_threshold, merging_threshold):
    for dilemma in dilemma_list : 
        update_contexts_with_dilemma(dilemma, Contexts, list_of_actions, adding_threshold)
        merge_similar_contexts(Contexts, merging_threshold)          
    return Contexts, len(Contexts['test']) 
```

# Grid Search Over Threshold Parameters with Performance Metrics Aggregation 

```python
# Define the threshold values

threshold_list_1 = np.linspace(0.01, 0.4, 21)  # List of candidate values for the first threshold parameter (adding threshold)
threshold_list_2 = np.linspace(0.01, 0.4, 21)  # List of candidate values for the second threshold parameter (merging threshold)
```

```python
R = {}  # Dictionary to store aggregated results for each N value

N_list = [30]      # List of sample sizes (number of distributions to generate per batch)
S_list = [1000]    # List of sample sizes for each distribution (number of samples per distribution)

# Calculate total number of iterations for tqdm progress bar (for all combinations and repetitions)
total_iterations = len(N_list) * len(S_list) * len(threshold_list_1) * len(threshold_list_2) * 5
print(total_iterations)

with tqdm(total=total_iterations, desc="Processing", unit="iter") as pbar:
    for i in N_list:
        for j in S_list:
            print(i, j)
            # Initialize lists to store results for each threshold combination
            # Results structure: [add_thresholds, merge_thresholds, mean_nb_contexts, mean_emd, mean_homogeneity, mean_loss]
            Results = [[], [], [], [], [], []]

            for t1 in threshold_list_1:
                for t2 in threshold_list_2:
                    # Initialize accumulators for metrics to average over 5 runs
                    mean_nb_contexts = 0
                    mean_loss = 0
                    mean_homog = 0
                    mean_emd = 0

                    # Repeat experiment 5 times for statistical stability
                    for _ in range(5):
                        # Generate filtered data distributions with current N and S
                        filtered_data = create_dist_list(i, [j])

                        # Compute clustering loss and metrics for current thresholds
                        interm_res = clustering_loss_3([t1, t2])
                        emd = interm_res[0]
                        nb_contexts = interm_res[1]
                        homog = interm_res[2]

                        # Custom combined loss: EMD + penalties for deviation in number of clusters and homogeneity
                        loss = emd + 0.6 * abs(5 - nb_contexts) + 0.6 * abs((1 / homog) - 1)

                        # Accumulate metrics for averaging
                        mean_emd += emd / 5
                        mean_nb_contexts += nb_contexts / 5
                        mean_homog += homog / 5
                        mean_loss += loss / 5

                        pbar.update(1)  # Update progress bar by one iteration

                    # Store averaged results for this threshold combination
                    Results[0].append(t1)
                    Results[1].append(t2)
                    Results[2].append(mean_nb_contexts)
                    Results[3].append(mean_emd)
                    Results[4].append(mean_homog)
                    Results[5].append(mean_loss)

            # Save the results keyed by sample size N
            R['N=' + str(i)] = Results.copy()

```

# Control Test: Generate Noisy Samples from a Single Base Distribution

```python
def create_dist_list_noisy_2(N, S_list, noise_level, base_distributions_list):
    """
    Generate a list of distributions by adding noise to base distributions.
    This function simulates samples from noisy versions of fixed base distributions,
    effectively creating only one 'context' per base distribution with slight variations.

    Parameters:
    - N: int, number of samples to generate per distribution size
    - S_list: list of ints, different sample sizes for generating samples
    - noise_level: float, uniform noise amplitude added to base distributions
    - base_distributions_list: list of keys to select base distributions from a global dictionary

    Returns:
    - distribution_dicts: list of dictionaries, each containing:
        'Action': base distribution name,
        'State': distribution name + sample size (context),
        'Reward': list of samples drawn according to the noisy distribution
    """
    distribution_dicts = []  # Initialize empty list to store generated distributions

    values = np.array([-1, 0, 1])  # Possible discrete sample values

    # Loop over each selected base distribution
    for dist in base_distributions_list:
        for i in range(len(S_list)):
            for j in range(N):
                distrib = base_distributions[dist]  # Retrieve base distribution probabilities
                
                # Add uniform noise to the base distribution to simulate variability
                noise = np.random.uniform(-noise_level, noise_level, size=distrib.shape)
                noisy_values = distrib + noise
                
                # Prevent negative probabilities and normalize to sum to 1
                noisy_values = np.maximum(noisy_values, 0)
                noisy_values /= np.sum(noisy_values)
                
                # Sample discrete values according to the noisy distribution
                samples = np.random.choice(values, size=S_list[i], p=noisy_values)
                
                # Append the sampled distribution to the list as a dictionary
                distribution_dicts.append({
                    'Action': dist,
                    'State': dist + ' ' + str(S_list[i]),  # Context label combining distribution and sample size
                    'Reward': samples.tolist()              # List of sampled values
                })

    random.shuffle(distribution_dicts)  # Shuffle to randomize order of samples
    
    return distribution_dicts


# ===============================
# Evaluate MBRL Agent on Noisy Sample List
# ===============================

noisy_list = create_dist_list_noisy_2(25, [1000], 0.1, base_distributions)  # Generate noisy data

classif = {}          # Initialize classification results dictionary
list_of_actions = []  # List to store actions encountered (can be used in MBRL_agent)

# Run the MBRL agent on the noisy sample list with given parameters
classif = MBRL_agent(noisy_list, classif, list_of_actions, 0.12, 0.03)


# ===============================
# Prepare Cluster Points Data for Visualization
# ===============================

cluster_points = []
for action, contexts in classif.items():
    for context_name, context_data in contexts.items():
        dist = context_data["Distribution"]  # Probability distribution of this cluster/context
        num_states = len(context_data.get("States", []))  # Number of associated states
        cluster_points.append({
            'Blame': dist.get(-1, 0),      # Probability mass on blame (-1)
            'Support': dist.get(1, 0),     # Probability mass on support (+1)
            'Action': action,
            'Context': context_name,
            'Color': 'r',                  # Placeholder color (can be customized)
            'Size': 120 + 80 * num_states  # Size scaled by number of states for visualization
        })
df_clusters = pd.DataFrame(cluster_points)  # Convert to DataFrame for plotting


# ===============================
# Prepare Dilemma Points Data for Visualization
# ===============================

colors = {
    'Negative bias': "#D7191C",  
    'Neutral': "#808080",  
    'Positive bias': "#1A9641",  
    'Extreme': "#2C7BB6",  
    'Balanced': "#FDAE61"   
}

dilemma_points = []
for d in noisy_list:
    dist = distribution(d['Reward'])  # Calculate empirical distribution from samples
    action = d["Action"]
    dilemma_points.append({
        'Blame': dist.get(-1, 0),     # Empirical blame probability
        'Support': dist.get(1, 0),    # Empirical support probability
        'Action': action,
        'Context': d.get("Context", "unknown"),
        'Color': colors[action]       # Color code based on action type
    })
df_dilemmas = pd.DataFrame(dilemma_points)


# ===============================
# Plot Scatter Plots of Dilemmas and Clustered Contexts by Action
# ===============================

actions = sorted(df_dilemmas["Action"].unique())

n_actions = len(actions)
ncols = 3
nrows = (n_actions + ncols - 1) // ncols  # Compute rows needed for subplot grid

fig, axes = plt.subplots(nrows, ncols, figsize=(6 * ncols, 5 * nrows), constrained_layout=True)
axes = axes.flatten()  # Flatten for easy indexing

for i, action in enumerate(actions):
    ax = axes[i]
    color = colors[action]

    # Plot dilemmas as scatter points
    subset_dilemmas = df_dilemmas[df_dilemmas["Action"] == action]
    ax.scatter(
        subset_dilemmas["Blame"],
        subset_dilemmas["Support"],
        s=50,
        color=color,
        alpha=0.9,
        edgecolors="black",
        label="Dilemmas"
    )

    # Plot cluster contexts as larger, semi-transparent squares
    subset_clusters = df_clusters[df_clusters["Action"] == action]
    ax.scatter(
        subset_clusters["Blame"],
        subset_clusters["Support"],
        s=subset_clusters["Size"],
        color=colors[action],
        alpha=0.5,
        edgecolors="black",
        marker='s',
        label="Contexts"
    )

    ax.set_title(f"{action}", fontsize=20, fontweight="bold")
    ax.set_xlim(-0.05, 1.05)
    ax.set_ylim(-0.05, 1.05)
    ax.set_xlabel("Blame", fontsize=20)
    ax.set_ylabel("Support", fontsize=20)
    ax.grid(True, linestyle="--", linewidth=0.5)

# Remove empty subplots if any
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.show()
```
